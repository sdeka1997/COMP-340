Matrix factorization works by breaking down the user-item matrix into the product of smaller matrices. I.e. given some matrix $X\in \mathbb{R}^{m\times n}$, we want to find matrices $U \in \mathbb{R}^{m\times k}$ and $V \in \mathbb{R}^{n\times k}$ such that $UV^T = X$. ## Question 13
- What is the difference between the k-means algorithm and the k-nearest-neighbor algorithm?

K-means is a clustering algorithm. It splits up a set of points into K clusters and tries to minimize the inter-point distances within each cluster. This algorithm qualifies as unsupervised ML because the points are not classified. 

K-nearest neighbors is a classification algorithm. To find the classification of a point, it combines the classification of the K nearest points. This algorithm qualifies as supervised because you are classifying points based on other already known classifications.

The ARIMA model is a combination of the AR (autoregressive) model and the MA (moving average) model. The regression is actually on the difference of the two.

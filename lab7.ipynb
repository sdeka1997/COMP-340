```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
plt.rcParams['figure.figsize']= (15,10)

airlines = pd.read_csv('airlines.csv')
airports = pd.read_csv('airports.csv')
planes = pd.read_csv('planes.csv')
weather = pd.read_csv('weather.csv')
flights = pd.read_csv('flights.csv')
```

## Characterizing seasonal patterns in departure delays
Are there any seasonal patterns in departure delays for flights from NYC?
To get a seasonal pattern in delays we will average departure delays for each month of the year from each of the three origin airports. 
Then we will bar plot  average departure delay against month for each of the three airports, and display them on the same plot.

```python
# your code here
flights.groupby(['month','origin'])['dep_delay'].mean().plot()
```

## Correlation between weather parameters at origin and departure delays

```python
# your code here
merged = flights.merge(weather, on=['origin','year','month', 'day', 'time_hour','hour'])
merged.head()
merged = merged[['dep_delay','temp','dewp','humid','wind_dir','wind_speed','precip','pressure','visib']]
merged.corr().dep_delay
```

## Relationship between departure delays and whether it is long-haul

```python
# your code here
flights['dep_delay'].corr(flights['distance'])
```

## Relationship between departure delays and size of aircraft

```python
# your code here
newmerge = pd.merge(flights, planes, how = 'inner', on = 'tailnum')
newmerge.groupby(['seats'])['dep_delay'].mean().plot(kind='bar')
```
